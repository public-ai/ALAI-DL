## 1. Merge

---

+ Merge란, 두 개 이상의 데이터프레임을 하나로 합치는 것을 의미합니다. <br>
+ **무엇을 기준**으로 합칠지를 결정해야 합니다.



### (1) inner Merge

inner merge는 `key`값을 기준으로 양쪽 모두 존재하는 것만을 가져오는 것을 의미합니다.

```python
pd.merge(author_df, reader_df,
         on="책", how='inner')
```

### (2) outer Merge

outer merge는 `key`값이 한쪽에만 있어도 합쳐줍니다. 이 때 없는 데이터는 `NaN`으로 처리합니다.

```python
df = pd.merge(author_df, reader_df, 
         on="책",how='outer')
df
```

*`NaN`은 해당 값이 결측되었음을 의미하는 값입니다.*<br>
*이를 제거하거나 치환하기 위한 메소드가 존재합니다.*

```python
# NaN가 있는 행 제거
df.dropna()
```

```python
# NaN가 있는 cell을 공백으로 치환
df.fillna(value="")
```

### (3) left & right Merge

left merge는 `key`값을 왼쪽의 데이터프레임 기준으로 잡고, <br>
right merge는 `key`값을 오른쪽의 데이터프레임 기준으로 잡습니다.

```python
pd.merge(author_df, reader_df, how='left')
```

```python
pd.merge(author_df, reader_df, how='right')
```

### (4) key column이름이 다른 경우

```python
reader_df.columns = ["독자","읽은 책"]
author_df.columns = ["저자","쓴 책"]
```

```python
reader_df.head(2)
```

```python
author_df.head(2)
```

각각의 데이터프레임 별로 키를 지정할 수 있다.

```python
pd.merge(author_df, reader_df, 
         left_on="쓴 책",right_on="읽은 책")
```

## 2. concat

---

+ concat란, 두 개의 데이터프레임을 행방향 혹은 열방향으로 합치는 것을 의미합니다. <br>
+ 어느 방향으로 합칠지를 결정해야 합니다.




### (1) 행방향

아래로 길게 붙입니다.

```python
# 예시 데이터셋
df1 = pd.DataFrame([81,82,83,80],columns=["score"])
df2 = pd.DataFrame([90,91,80,75],columns=["score"])
pd.concat([df1,df2],axis=0)
```

컬럼 이름을 기준으로 붙입니다. 컬럼 이름이 다른 경우, 결측 처리를 합니다.

```python
# 예시 데이터셋
df1 = pd.DataFrame([81,82,83,80],columns=["lang_score"])
df2 = pd.DataFrame([90,91,80,75],columns=["math_score"])
pd.concat([df1,df2],axis=0)
```

### (2) 열방향

옆으로 길게 붙입니다.

```python
# 예시 데이터셋
df1 = pd.DataFrame([81,82,83,80],columns=["score"])
df2 = pd.DataFrame([90,91,80,75],columns=["score"])
pd.concat([df1,df2],axis=1)
```

이름이 같더라도, 순차적으로 옆으로 붙입니다. 

```python
# 예시 데이터셋
df1 = pd.DataFrame([81,82,83,80],columns=["lang_score"])
df2 = pd.DataFrame([90,91,80,75],columns=["math_score"])
pd.concat([df1,df2],axis=1)
```

## 예제 데이터 ) 중간고사, 학생들의 국영수사과 성적표


학생들의 국영수사과 성적표를 Dataframe으로 저장한 후 다양하게 활용해보겠습니다.<br>

####  중간고사, 학생들의 국영수사과 성적표

|반   | 학생 이름 | 국어 | 영어 | 수학 | 사회 | 과학 |
|----|  ----   | --- |---| --- | --- | --- |
|1반 |승기 |80 |92 |70 | 65 | 92 |
|1반 |호동 |91 |75 |90 | 68 | 85 | 
|2반 |윤호 |86 |76 |42 | 72 | 88 |
|2반 |성진 |77 |92 |52 | 60 | 80 |
|3반 |서형 |75 |85 |85 | 92 | 95 |
|3반 |대근 |96 |90 |95 | 81 | 72 |
|4반 |준호 |91 |81 |92 | 81 | 73 |

```python
columns = ["반","이름","국어","영어","수학","사회","과학"]
scores = [["1반","승기",80,92,70,65,92],
         ["1반","호동",91,75,90,68,85],
         ["2반","윤호",86,76,42,72,88],
         ["2반","성진",77,92,52,60,80],
         ["3반","서형",75,85,85,92,95],
         ["3반","대근",96,90,95,81,72],
         ["4반","준호",91,81,92,81,73]]

df = pd.DataFrame(scores,columns=columns)
df
```

<br>

## 1. groupby

---

+ groupby 메소드는 데이터를 그룹 별로 분류하는 역할을 합니다. <br>
+ 연산 결과로 그룹 데이터를 나타내는 Groupby 클래스 객체를 반환합니다.



### (1) groupby 호출하기

```python
df.groupby(['반'])
```

groupby의 반환값은 `DataFrameGroupBy`입니다. 

이 객체는 for 구문을 통해 순회할 수 있습니다.

```python
for class_name, class_df in df.groupby(['반']):
    print("반 이름 : ",class_name)
    print(class_df)
    print('----\n')
```

```python
class_df # 4반의 dataframe
```

### (2) groupby 연산 적용하기

groupby 클래스는 그룹연산 메소드를 제공합니다. 

```python
# 반별 평균 성적 구하기
df.groupby("반").mean() 
```

```python
# 국어성적만 알고 싶으면, 아래와 같이 지정하면 됩니다. 
df.groupby("반")["국어"].mean()
```

이 외에도 아래와 같은 그룹연산들을 지원합니다.

| 메소드명 | 연산 |
| ----  | ---- |
| count | 그룹 데이터의 갯수 |
| mean  | 그룹 데이터 내 평균 |
| min   | 그룹 데이터 내 최소값|
| max   | 그룹 데이터 내 최대값|
| quantile | 그룹 데이터 내 분위수 |
| sum   | 그룹 데이터의 합 |
| std   | 그룹 데이터의 표준편차 |

## 2. pivot

---

+ 피봇이란 데이터 열 중에서 두 개의 열을 각각 행 인덱스, 열 인덱스로 사용하여 데이터를 조회하는 것을 의미합니다. <br>
+ 엑셀에서의 피봇과 유사한 기능으로 파악하면 됩니다.

```python
# 예제 데이터셋
# 전공이라는 이름의 열을 생성
df["전공"] = ["이과","이과","이과","문과","이과","문과","문과"]
df
```

### (1) 피봇 테이블 적용하기 

피봇하기 위해서 우리는 어떤 열을 index로 하고, 어떤 열을 columns으로 할지 정해야 합니다.

```python
df.pivot_table(values=["국어"],
               index=["반"],
               columns=["전공"])
```

### (2) 피봇의 집계함수 변경하기

기본적으로 집계함수는 평균값을 씁니다. 우리는 합, 갯수 등으로 pivot할 수 있습니다.

```python
df.pivot_table(values=["국어"],
               index=["반"],
               columns=["전공"],
               aggfunc="count")
```

### (3) 복수 개의  피봇 적용하기

```python
df.pivot_table(values=["국어","영어","수학"],
               index=["반"],
               columns=["전공"],
               aggfunc="count")
```
